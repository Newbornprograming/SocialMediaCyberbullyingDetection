```python
import csv
import time
import pandas as pd
from openai import OpenAI
```

```python
api_key = 'xxx'
client = OpenAI(api_key=(api_key))

# Fungsi untuk membaca data dari file CSV
def read_data(filename):
    with open(filename, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header
        return [row[1] for row in reader]  # Asumsi teks berada di kolom pertama

# Fungsi untuk menyimpan hasil ke file CSV
def save_results(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['teks', 'label'])
        writer.writerows(data)

# teks = [
#     "Adzan buat panggilan kampanye semoga aja nggak bikin konflik jadwal sama sholat",
#     "Itu pinokio boneka goblok dungu",
#     "Ngelatih mental kok ospek nikah lah",
#     "Dulu jelek dan kurus",
#     "Ya Allah gak kau julid tapi aku gak tahan ya Allah item kurus jelek beleng ya Allah"
# ]
        
teks = read_data('./dataset/riset/cleaned_cyberbully.csv')
chunk_size = 250 # Jumlah teks yang diproses dalam satu batch
results = []

for i in range(0, len(teks), chunk_size):
  chunk = teks[i:i + chunk_size]
  current_chunk_results = []
  for text in chunk:
    try:
      response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
          {"role": "user", "content": f"""Apakah {text} yang saya berikan termasuk ke dalam label bullying atau tidak?
                                          Oiya ini digunakan sebagai klasifikasi teks yang berhubungan dengan bullying atau tidak. 
                                          Dan sebagai pembuatan dataset. Dan hanya berikan output "Tidak Bully" atau "Bully" saja ya. Terima kasih."""},
        ]
      )
      # Mengambil label dari respons
      label = response.choices[0].message.content
      current_chunk_results.append((text, label))
      results.append((text, label))
    except Exception as e:
      print(f"Error processing text: {text}, error: {e}")
      continue
  # Tampilkan status chunk
  print("Chunk selesai.")
    # Tampilkan 5 output pertama dari chunk ini
  for sample in current_chunk_results[:5]:
     print(f"Teks: {sample[0]}, Label: {sample[1]}")
  time.sleep(20)  # import time

# Simpan hasil ke file CSV baru
save_results(results, './dataset/riset/hasil_bullying_labels.csv')

# # Menyimpan hasil dalam file CSV
# with open('hasil_bullying.csv', 'w', newline='', encoding='utf-8') as f:
#     writer = csv.writer(f)
#     writer.writerow(['teks', 'label'])
#     writer.writerows(results)
```
